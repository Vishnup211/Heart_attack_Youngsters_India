```python
ds['Gender'] = ds['Gender'].str.strip().str.title()
ds['Region'] = ds['Region'].str.strip().str.title()
ds['Smoking Status'] = ds['Smoking Status'].str.strip().str.title()
ds['Alcohol Consumption'] = ds['Alcohol Consumption'].str.strip().str.title()
ds['Physical Activity Level'] = ds['Physical Activity Level'].str.strip().str.title()
```

```python
print(ds['Gender'].unique())
```

```python
ds.head(5)
```

```python
ds.drop_duplicates(inplace=True)
```

```python
ds=ds.drop_duplicates(subset=['Age','Gender','Region'])
```

```python
ds['Risky Lifestyle'] = ds.apply(lambda row: 'Yes' if row['Smoking Status'] in ['Regularly', 'Occasionally'] or row['Alcohol Consumption'] in ['Regularly', 'Occasionally'] else 'No', axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of BMI
sns.histplot(ds['Blood Oxygen Levels (SpO2%)'], kde=True)
plt.show()

# Plot boxplot for Blood Pressure to check for outliers
sns.boxplot(x=ds['Blood Pressure (systolic/diastolic mmHg)'])
plt.show()
```

## To visualize the distribution of age

```python
sns.histplot(ds['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Gender', data=ds)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Maximum Heart Rate Achieved', data=ds)
plt.title('Age vs Maximum Heart Rate Achieved')
plt.show()
```

## Heart Attack Likelihood across different Regions or Smoking Status

```python
pd.crosstab(ds['Region'], ds['Heart Attack Likelihood']).plot(kind='bar', stacked=True)
plt.title('Heart Attack Likelihood by Region')
plt.ylabel('Count')
plt.show()
```

```python
ds['Smoking Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.title('Smoking Status Proportion')
plt.ylabel('')
plt.show()
```
